```python
#library
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

#load dataset
file_path = r"D:\feliz navidad\TUGAS KULIAH\DATA MINING FILE\TUGAS 4\DatasetpProduksiPadi.xlsx"
excel_data = pd.ExcelFile(file_path)

print(excel_data.sheet_names)
```

```python
# Load data dari 'data training'
df = excel_data.parse('data training')
```

```python
# Data Cleaning
df_cleaned = df[['Tahun', 'Produksi Padi(Ton)']].dropna()
df_cleaned.columns = ['Tahun', 'Produksi_Padi']
df_cleaned['Tahun'] = df_cleaned['Tahun'].astype(int)
df_cleaned['Produksi_Padi'] = df_cleaned['Produksi_Padi'].astype(float)
```

```python
df_cleaned.isnull().sum()
```

```python
#fitur lagging
df_cleaned['Produksi_Padi_Lag1'] = df_cleaned['Produksi_Padi'].shift(1)
df_cleaned = df_cleaned.dropna()
```

```python
df_cleaned.head()
```

```python
# memisahkan fitur (X) dan target (Y)

#fitur produksi thn sebelumnya
X = df_cleaned['Produksi_Padi_Lag1']

#target produksi padi
Y = df_cleaned['Produksi_Padi']
```

```python
X.head()
```

```python
Y.head()
```

```python
#split data ke data train dan data test dengan perbandingan 70 : 30
train_size = int(len(df_cleaned) * 0.7)
train_data = df_cleaned[:train_size]
test_data = df_cleaned[train_size:]
```

```python
 #fitur dan target untuk training and testing
X_train = train_data[['Produksi_Padi_Lag1']]
Y_train = train_data['Produksi_Padi']
X_test = test_data[['Produksi_Padi_Lag1']]
Y_test = test_data['Produksi_Padi']
```

```python
print("Ukuran X_train:", X_train.shape)
print("Ukuran Y_train:", Y_train.shape)
print("Ukuran X_test:", X_test.shape)
print("Ukuran Y_test:", Y_test.shape)
```

```python
# modeling menggunakan Linier Regression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
#prediksi 
y_pred =  model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)
```

```python

# Mendapatkan nilai terakhir dari data
latest_value = df_cleaned['Produksi_Padi'].iloc[-1]

# Membuat prediksi untuk 2023
latest_value_df = pd.DataFrame({'Produksi_Padi_Lag1': [latest_value]})
pred_2023 = model.predict(latest_value_df)

# Membuat prediksi untuk 2024
pred_2024_df = pd.DataFrame({'Produksi_Padi_Lag1': [pred_2023[0]]})
pred_2024 = model.predict(pred_2024_df)
```

```python
evaluation_results = {
    'MSE': mse,
    'RMSE': rmse,
    'MAPE (%)': mape * 100,
    'Prediksi 2023': pred_2023[0],
    'Prediksi 2024': pred_2024[0]
}
```

```python
#menampilkan hasil evaluasi
print("Hasil Evaluasi Model:")
print(f"MSE (Mean Squared Error): {evaluation_results['MSE']:.2f}")
print(f"RMSE (Root Mean Squared Error): {evaluation_results['RMSE']:.2f}")
print(f"MAPE (Mean Absolute Percentage Error): {evaluation_results['MAPE (%)']:.2f}%")
print(f"Prediksi Produksi Padi 2023: {evaluation_results['Prediksi 2023']:.2f} ton")
print(f"Prediksi Produksi Padi 2024: {evaluation_results['Prediksi 2024']:.2f} ton")
```
